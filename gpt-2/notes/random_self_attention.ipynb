```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

np.random.seed(42)

input_words = ["Anna", "bought", "a", "red", "and", "a", "green", "apple"]
output_words = ["Anna", "bought", "a", "red", "and", "a", "green", "apple"]

attention_matrix = np.random.rand(len(output_words), len(input_words))

plt.figure(figsize=(10, 8))
ax = sns.heatmap(attention_matrix, xticklabels=input_words, yticklabels=output_words, cmap="Oranges", cbar_kws={'label': 'Attention'})

ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

plt.show()
```
