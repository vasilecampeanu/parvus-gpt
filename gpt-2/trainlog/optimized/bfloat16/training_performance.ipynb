```python
import matplotlib.pyplot as plt
```

```python
devices = ['CUDA (A100 SMX) Float32 High', 'CUDA (A100 SMX) BFloat16']
times = [8.93, 8.30]
```

# Unoptimized training performance

```python
import matplotlib.pyplot as plt

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(devices, times, color=['blue', 'green'])

# Annotate the bars with the exact times
for bar, time in zip(bars, times):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, f'{time:.2f}', ha='center', va='bottom')

# Set labels and title
plt.xlabel('Device')
plt.ylabel('Average Training Time (seconds)')
plt.title('Average Training Time by Device')

# Display the plot
plt.show()
```

```python
# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(devices, times, color=['blue', 'green'])

# Annotate the bars with the exact times
for bar, time in zip(bars, times):
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, f'{time:.2f}', va='center', ha='left')

# Set labels and title
plt.ylabel('Device')
plt.xlabel('Average Training Time (seconds)')
plt.title('Average Training Time by Device')

# Display the plot
plt.show()
```
