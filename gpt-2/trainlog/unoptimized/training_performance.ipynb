```python
import matplotlib.pyplot as plt
```

```python
# Data for the training times
devices = ['CPU', 'MPS', 'CUDA (RTX 4090)', 'CUDA (A100 SMX)']
times = [484.71, 116, 15.76, 26.22]
```

# Unoptimized training performance

```python
# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(devices, times, color=['blue', 'green', 'red'])
plt.xlabel('Device')
plt.ylabel('Average Training Time (seconds)')
plt.title('Average Training Time by Device')
plt.yscale('log')  # Using logarithmic scale for better visualization

# Annotate the bars with the exact times
for i, time in enumerate(times):
    plt.text(i, time + 5, f'{time:.2f}', ha='center', va='bottom')

plt.show()
```

```python
# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(devices, times, color=['blue', 'green', 'red', 'purple'])
plt.ylabel('Device')
plt.xlabel('Average Training Time (seconds)')
plt.title('Average Training Time by Device')
plt.xscale('log')  # Using logarithmic scale for better visualization

# Annotate the bars with the exact times
for i, time in enumerate(times):
    plt.text(time + 5, i, f'{time:.2f}', va='center', ha='left')

plt.show()
```
